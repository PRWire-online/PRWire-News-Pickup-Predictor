## What This Notebook Does

- Loads the PRWire benchmark dataset (20 press release campaigns)
- Visualizes pickup scores across tiers
- Shows average syndication reach by distribution tier
- Demonstrates the pickup predictor function

```python
# Install dependencies
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = (10, 5)

print('Libraries loaded successfully')
```

```python
# Load benchmark dataset
df = pd.read_csv('../dataset/reach_benchmarks.csv')
print('Dataset shape:', df.shape)
print(df.head(10))
```

```python
# Pickup Score by Campaign
plt.figure(figsize=(12, 5))
colors = {'local': '#6366f1', 'national': '#10b981', 'global': '#f59e0b'}
bar_colors = df['tier'].map(colors)
plt.bar(df['id'], df['pickup_score'], color=bar_colors)
plt.title('PRWire — Pickup Score by Press Release Campaign', fontsize=14, fontweight='bold')
plt.xlabel('Campaign ID')
plt.ylabel('Pickup Score (0-100)')
plt.xticks(df['id'])
plt.axhline(y=70, color='red', linestyle='--', alpha=0.5, label='High pickup threshold (70)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Average Reach by Tier
tier_avg = df.groupby('tier')['estimated_reach'].mean().reset_index()
tier_avg.columns = ['Tier', 'Average Reach']

plt.figure(figsize=(7, 4))
plt.bar(tier_avg['Tier'], tier_avg['Average Reach'], color=['#6366f1', '#f59e0b', '#10b981'])
plt.title('PRWire — Average Syndication Reach by Tier', fontsize=14, fontweight='bold')
plt.xlabel('Distribution Tier')
plt.ylabel('Average Estimated Reach')
plt.tight_layout()
plt.show()

print(tier_avg.to_string(index=False))
```

```python
# Pickup Predictor Function
TIER_MULTIPLIERS = {'local': 0.4, 'national': 0.7, 'global': 1.0}
BASE_REACH = {'local': 5000, 'national': 20000, 'global': 50000}

def predict_pickup(press_releases: int, tier: str = 'global') -> dict:
    tier = tier.lower()
    multiplier = TIER_MULTIPLIERS.get(tier, 1.0)
    pickup_score = min(100, round(40 + press_releases * multiplier))
    estimated_reach = round(BASE_REACH.get(tier, 50000) + press_releases * multiplier * 1000)
    return {
        'press_releases': press_releases,
        'tier': tier,
        'pickup_score': pickup_score,
        'estimated_reach': estimated_reach
    }

# Test examples
examples = [
    (10, 'local'),
    (25, 'national'),
    (50, 'global'),
    (100, 'global'),
]

print(f"{'PRs':<8} {'Tier':<12} {'Score':<10} {'Reach'}")
print('-' * 45)
for pr, tier in examples:
    r = predict_pickup(pr, tier)
    print(f"{r['press_releases']:<8} {r['tier']:<12} {r['pickup_score']:<10} ~{r['estimated_reach']:,}")
```
